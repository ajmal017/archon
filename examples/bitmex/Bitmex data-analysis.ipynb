```python
import pandas as pd
```

```python
df = pd.read_csv("bitmex_candles1.csv")
```

```python
print ("candles ", len(df))
```

```python
df['change'] = df['close'].diff()
df['roc'] = df['change']/df['close']
df['Quantile_rank']=pd.qcut(df['roc'],4,labels=False)
```
